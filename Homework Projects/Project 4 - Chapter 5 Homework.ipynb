### 3. Create a dictionary where each key points to a unique numpy array. All arrays should be of the same length. Then transform that dictionary into a pandas dataframe.

```python
unique_dict = {"AAPL Rev(in millions)":{},
              "AAPL Market Cap(in millions)": {},
              "AAPL COGS(in millions)": {},
              }
for key in unique_dict:
    for i in range(2005,2021):
        unique_dict[key][i] = np.random.random()*1000

unique_df = pd.DataFrame(unique_dict)
unique_df["P/S Ratio"] = unique_df["AAPL Market Cap(in millions)"]/unique_df["AAPL Rev(in millions)"]
unique_df
```

### 7. Create a function that plots 2 to 3 different variables from the dataframe. The function may look something like "def plot_data(df,names, etc...). Each plot should include a legend and title.

```python
for key in tsla_inc_stat:
    fig, ax = plt.subplots(figsize = (20,10))
    tsla_inc_stat[key].plot.line(color = "C0", ax = ax)
    tsla_inc_stat[key].plot.line(ls="", marker="*", ax=ax)
    plt.title(key, fontsize = 25)
    plt.show()
    plt.close()
```

### 8. Create a bar graph plot using the data you have imported.

```python
for key in tsla_inc_stat:
    fig, ax = plt.subplots(figsize = (20,10))
    tsla_inc_stat[key].plot.bar(color = "c0", ax = ax)
    tsla_inc_stat[key].plot.bar(ls="", marker="*", ax=ax)
    plt.title(key, fontsize = 25)
    plt.show()
    plt.close()
```
