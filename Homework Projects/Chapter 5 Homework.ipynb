### 3. Create a dictionary where each key points to a unique numpy array. All arrays should be of the same length. Then transform that dictionary into a pandas dataframe.

```python
unique_dict = {"AAPL Rev(in millions)":{},
              "AAPL Market Cap(in millions)": {},
              "AAPL COGS(in millions)": {},
              }
for key in unique_dict:
    for i in range(2005,2021):
        unique_dict[key][i] = np.random.random()*1000

unique_df = pd.DataFrame(unique_dict)
unique_df["P/S Ratio"] = unique_df["AAPL Market Cap(in millions)"]/unique_df["AAPL Rev(in millions)"]
unique_df
```
