```python
import pandas as pd
import pandas_datareader as web
import datetime 
import numpy
from mylib.mean_variance import *
from datlib.FRED import *
start = datetime.datetime(1970,1,1)
end = datetime.datetime.today()
```

```python
data_codes = {"Net Farm Income": "B1448C1A027NBEA"}
tickers = ["DE"]
farm_data = gather_data(data_codes, start = start, end = end, freq = "A")
farm_data["DE Yearly Price Change"] = web.DataReader("DE", 
                                                     "yahoo", 
                                                     start = start, 
                                                     end = end)["Adj Close"].resample("Y").last().pct_change()

farm_data["Net Farm Income % Change"] = farm_data["Net Farm Income"].pct_change()
farm_data["DE Yearly Price Change Lag"] = farm_data["DE Yearly Price Change"].shift()
farm_data = farm_data.dropna()
farm_data
```

```python
farm_ols = ols(farm_data, x_vars = ["Net Farm Income % Change"], y_var = ["DE Yearly Price Change Lag"])
farm_ols
```
