```python
import pandas as pd
import numpy as np
import pandas_datareader as web
import matplotlib.pyplot as plt
import datetime
#imported a library that I created so that I wouldn't clutter this notebook
from mylib.mean_variance import *
import statsmodels.api as sm
from datlib.FRED import *
from datlib.plots import *
benchmark = "^GSPC"
plt.rcParams.update({"font.size":40})
start = datetime.datetime(2011,1,1)
end = datetime.datetime.today()
rf = web.DataReader("^TNX", "yahoo", start = datetime.datetime(2021,10,1), end = end)["Adj Close"]
rf = pd.DataFrame(rf)
ticker_lst = ["XLK", "XLY", "XLRE", "XLV", "XLP", "XLE", "XLF", "XLI", "XLC", "BND", "SUB"]
```

```python
portfolios = mean_variance(ticker_lst, 
                           num_ports = 50000, 
                           start = start, 
                           end = end, 
                           benchmark = benchmark,
                           rf = rf["Adj Close"][-1])
portfolios
```
