## Create an indicator variable

```python
def create_indicator_variable(data, 
                              indicator_name, 
                              index_name, 
                              target_index_list):
    # prepare a column with the name of the indicator variable
    data[indicator_name] = 0
    # for each index whose name matches the entry in target_index_list
    # a 1 will be recorded
    for index in target_index_list:
        # check if ISO code of row matches the ISO Code from the target index list
        # set value of the indicator name (column) to 1 for all rows
        data.loc[index, [indicator_name]] = round(int(1),0)
        
data = fraser_data
index_name = data.index.names[0]
indicator_names = "North America"
# Cuba, Grenada, Saint Kitts, Saint Lucia, Saint Vincent are missing
countries_in_north_america = ["BHS", "BRB", "BLZ", "CAN", 
                              "CRI", "DOM", "SLV", "GTM", 
                              "HTI", "HND", "JAM", "MEX", 
                              "NIC", "PAN", "TTO", "USA"] 
create_indicator_variable(data = data, 
                          indicator_name = indicator_names, 
                          index_name = index_name, 
                          target_index_list = countries_in_north_america)       
```

## Continue

```python
def create_quantile(n, data, year, quantile_var, quantile_name):
    # year becuase we have panel data and we want to look at that given year
    year_index = data.index.get_level_values("Year") == year
    year_data = data[year_index][quantile_var]
    # select the subset of the data frame by year and variable of interest
    quantile_values_dct = {i:year_data.quantile(i/n) for i in range(1, n+1)}
    for index in data[year_index].index:
        # identify value of the variable of interest
        val = data.loc[index][quantile_var]
        # compare that value to the values that divide each quantile
        for i in range(1, n+1):
            if val <= quantile_values_dct[i]:
                data.loc[index, [quantile_name]] = int((n+1) - i)
                break
            else:
                continue
years = data.groupby("Year").mean().index
for year in years:
    create_quantile(n, 
                   data, 
                   year, 
                   quantile_var,   
                   quantile_name)
```
