# Advanced Data Analysis

```python
import pandas as pd
import pandas_datareader as web
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_excel("mpd2018.xlsx", 
                     sheet_name = "Full data", 
                     index_col = [0,2])
data
```

#### list of years

```python
years = list(set(data.index.get_level_values("year")))
years
```

#### Mean of every country's data

```python
data.groupby('countrycode').mean()
```

#### List of countries

```python
countries = list(data.groupby('countrycode').mean().index)
countries
```

#### Plot data by country

```python
pairs = [("USA", "CAN"), 
         ("VEN", "MEX"), 
         ("NOR", "FIN", "DNK", "NLD")]
linestyles = ["-", ":", "--", "-."]
```

```python
plt.rcParams.update({"font.size": 32})
plt.rcParams.update({"legend.fontsize": 25, 
                     "legend.handlelength": 2})
for pair in pairs:
    fig, ax = plt.subplots(figsize = (20,10))
    for i in range(len(pair)):
        country = pair[i]
        linestyle = linestyles[i]
        data.loc[country, :]["cgdppc"].dropna().plot.line(ax = ax,
                                                          label = country,
                                                          linestyle = linestyle)
    plt.xlim([1825, max(years)])
    plt.ylabel("$Real$ $GDP$ $Per$ $Capita$\n2011 U.S. Dollars")
    plt.legend()
    plt.show()
    plt.close()
```

## Create an indicator variable

```python
def create_indicator_variable(data, 
                              indicator_name, 
                              index_name, 
                              target_index_list):
    # prepare a column with the name of the indicator variable
    data[indicator_name] = 0
    # for each index whose name matches the entry in target_index_list
    # a 1 will be recorded
    for index in target_index_list:
        # check if ISO code of row matches the ISO Code from the target index list
        # set value of the indicator name (column) to 1 for all rows
        data.loc[index, [indicator_name]] = round(int(1),0)
        
data = fraser_data
index_name = data.index.names[0]
indicator_names = "North America"
# Cuba, Grenada, Saint Kitts, Saint Lucia, Saint Vincent are missing
countries_in_north_america = ["BHS", "BRB", "BLZ", "CAN", 
                              "CRI", "DOM", "SLV", "GTM", 
                              "HTI", "HND", "JAM", "MEX", 
                              "NIC", "PAN", "TTO", "USA"] 
create_indicator_variable(data = data, 
                          indicator_name = indicator_names, 
                          index_name = index_name, 
                          target_index_list = countries_in_north_america)       
```

#### Ploting Countries

```python
import datetime
year = 2009
plot_data = data[data.index.get_level_values("Year") == year]
fig, ax = plt.subplots(figsize = (24,12))
plot_data.plot.scatter(x = "EFW", 
                       y = "RGDP Per Capita", 
                       c = "North America", 
                       cmap = "coolwarm", 
                       ax = ax, 
                       s = 55)
ax.set_title(str(year), fontsize = 50)
plt.show()
plt.close()
```
